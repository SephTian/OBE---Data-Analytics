
**OPT PARAMETER** (Joseph)

```python
from sklearn.metrics import silhouette_score as ss
import itertools
epsilons = [0.2,0.4,0.5,0.7,0.8,0.9,1,1.1,1.2]
min_samples = [1,3,5,7,9,11,13,15]

parameter = list(itertools.product(epsilons, min_samples))
def opt_para_dbscan(combinations, X):
  scores = []
  all_labels_list = []

  for i, (eps, num_samples) in enumerate(combinations):
    dbscan_cluster_model = DBSCAN(eps=eps, min_samples=num_samples).fit(X)
    labels = dbscan_cluster_model.labels_
    labels_set = set(labels)
    num_clusters = len(labels_set)
    if -1 in labels_set:
      num_clusters -= 1

    if (num_clusters < 2) or (num_clusters > 7):
      scores.append(-100)
      all_labels_list.append('end')
      continue

    scores.append(ss(X, labels))
    all_labels_list.append(labels)

  best_index = np.argmax(scores)
  best_parameters = combinations[best_index]
  best_labels = all_labels_list[best_index]
  best_score = scores[best_index]

  return {'best_epsilon': best_parameters[0],
          'best_min_samples': best_parameters[1],
          'best_labels': best_labels,
          'best_score': best_score}

best_dict = opt_para_dbscan(parameter, x)
```

Density based parameter search (Stanley)

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn import metrics

def optimize_dbscan_density_search(X, min_eps, max_eps, min_samples):
    best_score = -1
    best_dbscan = None

    # Compute the k-distance graph
    nn = NearestNeighbors(n_neighbors=min_samples)
    nn.fit(X)
    distances, _ = nn.kneighbors(X)
    k_distance = distances[:, -1]

    # Search for optimal parameters within the specified range
    for eps in range(int(min_eps * 100), int(max_eps * 100) + 1):
        eps /= 100
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        dbscan.fit(X)
        labels = dbscan.labels_
        unique_labels = len(set(labels))
        if unique_labels < 2 or unique_labels > 7:
            # Skip combinations with less than two unique cluster labels
            continue
        score = metrics.silhouette_score(X, labels)
        if score > best_score:
            best_score = score
            best_dbscan = dbscan
    print('Best Score ', best_score)
    return best_dbscan

min_eps = 0.4  # Minimum epsilon value to search
max_eps = 1.9  # Maximum epsilon value to search
min_samples = 7  # Minimum number of samples for core point

# Optimize DBSCAN using density-based search
best_dbscan = optimize_dbscan_density_search(df_kualitas_individu_thd_obe, min_eps, max_eps, min_samples)
```

Density peak selection (Stanley)

```python
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
from sklearn import metrics

def optimize_dbscan_density_peak(X, k, delta):
    nn = NearestNeighbors(n_neighbors=k)
    nn.fit(X)
    distances, _ = nn.kneighbors(X)
    density = 1 / distances[:, -1]
    density_sorted_indices = np.argsort(density)[::-1]

    for i in range(len(X)):
        if density[density_sorted_indices[i]] < delta:
            break

    eps = distances[density_sorted_indices[i - 1], -1]
    dbscan = DBSCAN(eps=eps, min_samples=k)
    dbscan.fit(X)

    # Compute the silhouette score
    labels = dbscan.labels_
    score = metrics.silhouette_score(X, labels)

    return dbscan, score

k = 7  # Number of nearest neighbors for density calculation
delta = 0.1  # Threshold for density peak selection
best_density, score = optimize_dbscan_density_peak(df_kualitas_individu_thd_obe, k, delta)

print("DBSCAN Score:", score)
```
